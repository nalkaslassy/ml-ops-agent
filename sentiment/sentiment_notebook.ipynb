```python
import sys
sys.path.insert(0, '..')

import pandas as pd
import numpy as np
import warnings
warnings.filterwarnings('ignore')

from sentiment.sentiment_model import SentimentModel
```

## Load Data

Replace the sample data below with your own CSV. Only two things to change:
- `TEXT_COL` — the column that contains the text (reviews, tickets, responses, etc.)
- `TARGET_COL` — the column that contains the label (positive/negative, 1/0, etc.)

```python
TEXT_COL   = 'review'
TARGET_COL = 'sentiment'

# Replace with: df = pd.read_csv('your_file.csv')
data = [
    ("This product is absolutely amazing, I love it!", 1),
    ("Terrible experience, would not recommend to anyone.", 0),
    ("Great customer service, very helpful and fast.", 1),
    ("The product broke after one week. Very disappointed.", 0),
    ("Exceeded my expectations, will definitely buy again.", 1),
    ("Worst purchase I have ever made, complete waste of money.", 0),
    ("Outstanding quality and delivered ahead of schedule.", 1),
    ("Customer support was rude and unhelpful.", 0),
    ("Fantastic value for the price, highly recommended.", 1),
    ("Product looks nothing like the photos online.", 0),
    ("Very happy with my purchase, works perfectly.", 1),
    ("Took forever to arrive and packaging was damaged.", 0),
    ("Solid build quality, exactly what I needed.", 1),
    ("Do not buy this. It stopped working immediately.", 0),
    ("Five stars all the way, exceptional experience.", 1),
    ("Overpriced and underwhelming, not worth it.", 0),
    ("Works as described, no complaints at all.", 1),
    ("Received wrong item and return process was a nightmare.", 0),
    ("Love this product, using it every single day.", 1),
    ("Broke on first use. Absolute garbage.", 0),
] * 30  # repeat to get enough rows for training

df = pd.DataFrame(data, columns=[TEXT_COL, TARGET_COL])
df = df.sample(frac=1, random_state=42).reset_index(drop=True)

print(f'Rows: {len(df)} | Positive rate: {df[TARGET_COL].mean():.1%}')
df.head()
```

## Train the Model

Tests all 3 algorithms, tunes each one, picks the best.

```python
model = SentimentModel(
    text_col=TEXT_COL,
    target_col=TARGET_COL,
    label_names=['Negative', 'Positive']
)
model.fit(df)
```

## Predict on New Text

Pass in any list of strings — reviews, support tickets, survey responses, whatever.

```python
new_texts = [
    "This is the best thing I have ever bought!",
    "Completely broken on arrival, very frustrating.",
    "Decent product, does what it says.",
    "Fast shipping and great packaging, very happy.",
    "Never buying from this company again."
]

model.predict(new_texts)
```
